```python
list_of_twelves = []
list_of_not_twelves = []
col_00, col_01, col_02, col_03, col_04, col_05, col_06, col_07, col_08, col_09, col_10, col_11, col_12 = [],[],[],[],[],[],[],[],[],[],[],[],[]
for rows in cog_20_cog[0]:
    test = rows
    if test.count(',') == 12:
        list_of_twelves.append(rows)
    else:
        list_of_not_twelves.append(rows)
        
    col_00.append(test.split(',')[0])
    col_01.append(test.split(',')[1])
    col_02.append(test.split(',')[2])
    col_03.append(test.split(',')[3])
    col_04.append(test.split(',')[4])
    col_05.append(test.split(',')[5])
    col_06.append(test.split(',')[6])
    col_07.append(test.split(',')[7])
    col_08.append(test.split(',')[8])
    col_09.append(test.split(',')[9])
    col_10.append(test.split(',')[10])
    col_11.append(test.split(',')[11])
    col_12.append(test.split(',')[12])
    
cog_20_cog[0] = col_00
cog_20_cog[1] = col_01
cog_20_cog[2] = col_02
cog_20_cog[3] = col_03
cog_20_cog[4] = col_04
cog_20_cog[5] = col_05
cog_20_cog[6] = col_06
cog_20_cog[7] = col_07
cog_20_cog[8] = col_08
cog_20_cog[9] = col_09
cog_20_cog[10] = col_10
cog_20_cog[11] = col_11
cog_20_cog[12] = col_12
```
