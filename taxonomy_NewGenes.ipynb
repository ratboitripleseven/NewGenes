# Checking the taxonomies

## NCBI

NCBI has variable taxonomy!

```python
newgenes_NCBI_taxo = taxonomy_table_processed.drop(columns=["Genome size(Mb)", "GC(%)", "GenBank accession number", "BioProject", "link", "GDTB_taxonomy" ])
newgenes_NCBI_taxo["phylum"]= [x.split(';')[1] for x in newgenes_NCBI_taxo["Taxonomy"].values]
newgenes_NCBI_taxo["class"]= [x.split(';')[2] if len(x.split(';'))>2 else None  for x in newgenes_NCBI_taxo["Taxonomy"].values]
newgenes_NCBI_taxo["order"]= [x.split(';')[3] if len(x.split(';'))>3 else None  for x in newgenes_NCBI_taxo["Taxonomy"].values]
```

```python
newgenes_NCBI_taxo
```

nones found in order is a subset of nones in class

```python
newgenes_NCBI_taxo[newgenes_NCBI_taxo['class'].isnull().values]
```

```python
newgenes_NCBI_taxo[newgenes_NCBI_taxo['order'].isnull().values]
```

```python
newgenes_NCBI_taxo.drop(newgenes_NCBI_taxo[newgenes_NCBI_taxo['order'].isnull().values].index, inplace=True)
```

```python
newgenes_NCBI_taxo
```

## class distribution per phylum

```python
fig = plt.figure(figsize=(30,50))
for i,phylum in enumerate(newgenes_NCBI_taxo["phylum"].unique()):
    ax = fig.add_subplot(11, 4, i+1)
    current_phylum_table = newgenes_NCBI_taxo[newgenes_NCBI_taxo["phylum"] == phylum]
    
    classes_per_current_phylum_table = current_phylum_table["class"].unique()
    count_per_class_per_current_phylum = []
    for k in classes_per_current_phylum_table:
        count_per_class_per_current_phylum.append(len(current_phylum_table[current_phylum_table["class"] == k]))

    ax.set_title(phylum)
    ax.bar(list(classes_per_current_phylum_table), count_per_class_per_current_phylum)
```

```python
max_length = 0
index = ""
for taxonomy in taxonomy_table_processed["Taxonomy"]:
    tax_list = taxonomy.split(";")
    if len(tax_list)>max_length:
        max_length = len(tax_list)
        index = taxonomy 

max_length
index
```

```python
newgenes_GTDB_taxo = taxonomy_table_processed.drop(columns=["Genome size(Mb)", "GC(%)", "GenBank accession number", "BioProject", "link", "Taxonomy" ])
```

```python
newgenes_GTDB_taxo["phylum"]= [x.split(';')[1] for x in newgenes_GTDB_taxo["GDTB_taxonomy"].values]
newgenes_GTDB_taxo["class"]= [x.split(';')[2] for x in newgenes_GTDB_taxo["GDTB_taxonomy"].values]
newgenes_GTDB_taxo["order"]= [x.split(';')[3] for x in newgenes_GTDB_taxo["GDTB_taxonomy"].values]
newgenes_GTDB_taxo["family"]= [x.split(';')[4] for x in newgenes_GTDB_taxo["GDTB_taxonomy"].values]
newgenes_GTDB_taxo["genus"]= [x.split(';')[5] for x in newgenes_GTDB_taxo["GDTB_taxonomy"].values]
newgenes_GTDB_taxo["species"]= [x.split(';')[6] for x in newgenes_GTDB_taxo["GDTB_taxonomy"].values]
```

## uniqe phylums

```python
stats = []
for phylum in newgenes_GTDB_taxo["phylum"].unique():
    amount = len(newgenes_GTDB_taxo[newgenes_GTDB_taxo["phylum"] == phylum])
    print(f'phylum: {phylum}, count {amount}')
    stats.append(amount)
    
```

```python
plt.figure(figsize=(30,18))
plt.bar(newgenes_GTDB_taxo["phylum"].unique(), stats)
```

## unique class per phylum

```python
stats_class_per_phylum = []
for phylum in newgenes_GTDB_taxo["phylum"].unique():
    phylum_table = newgenes_GTDB_taxo[newgenes_GTDB_taxo["phylum"] == phylum]
    unique_classes_per_phylum = len(phylum_table["class"].unique())
    print(f'phylum: {phylum}, count {unique_classes_per_phylum}')
    stats_class_per_phylum.append(unique_classes_per_phylum)
```

```python
plt.figure(figsize=(30,18))
plt.bar(newgenes_GTDB_taxo["phylum"].unique(), stats_class_per_phylum)
```
