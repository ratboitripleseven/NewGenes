```python
# sanity check for preprocessed data!
for csv_file in os.listdir(PATH_TO_PREPROCESSED_HGTDB_FILES):
    df_1 = pd.read_csv(os.path.join(PATH_TO_PREPROCESSED_HGTDB_FILES, csv_file), index_col='ID')
    df_2 = pd.read_csv(os.path.join(PATH_TO_PREPROCESSED_HGTDB_FILES_2, csv_file), index_col='ID')
    
    if len(df_1) != len(df_2):
        print(f'len of 1 {len(df_1)} and len of 2 {len(df_2)}')
    
```

```python
def check_extraneity(dataframe_location):
    sim1 = dataframe_location['Sim1']
    sim2 = dataframe_location['Sim2']
    sim3 = dataframe_location['Sim3']
    simt = dataframe_location['SimT']
    if sim1 * sim3 > 0 and (abs(sim1)>= 1 or abs(sim3)>= 1):
        #total_extraneous = total_extraneous + 1
        extraneous = True
    elif abs(simt) >= 1:
        #total_extraneous = total_extraneous + 1
        extraneous = True
    else:
        extraneous = False
    return extraneous
```

```python
def drop_df(dataframe):
    dataframe.dropna(inplace=True)
    if dataframe.isna().sum().sum() >0:
        print(df[df.isna().any(axis=1)])
    dataframe = dataframe.bfill(axis='columns')
    if dataframe.isna().sum().sum() >0:
        print(df[df.isna().any(axis=1)])
    return dataframe
```

```python
to_csv = []
for csv_file in os.listdir(PATH_TO_PREPROCESSED_HGTDB_FILES):
    #df = pd.read_csv(os.path.join(PATH_TO_PREPROCESSED_HGTDB_FILES, csv_file), index_col='ID')
    df = pd.read_csv(os.path.join(PATH_TO_PREPROCESSED_HGTDB_FILES_2, csv_file), index_col='ID')
    #df_temp_data = pd.read_csv(os.path.join(TEMP_DATA, csv_file.replace('.csv','.tsv')), sep='\t', index_col="Synonym")
    
    df = df.drop(columns=["FunctionCode","Strand","AADev","Length","GC1","GC2","GC3","GCT","SD1","SD2","SD3","SDT","Mah"])
    
    df = drop_df(df)
    df['extraneous'] = [check_extraneity(df.iloc[i]) for i in range(len(df))]
    #HGT check
    hgt_check = df['HGT'] == 1
    if (hgt_check[hgt_check==False].count() + hgt_check[hgt_check==True].count()) == len(df):
        print('okg')
    
    
    
    print(f'{csv_file}')
    #print(f'lenght before {len(df_temp_data)} lenght after {len(df_temp_data)}')
    print(f"Total hgt counted { df[df['HGT']==1]['HGT'].count()}")
    print(f"Total extraneous counted {df[df['extraneous'] == True]['extraneous'].count()}")
    print(f"Total overlap {df[(df['HGT']==1) & (df['extraneous'] == True)]['HGT'].count()}")
    
    #['genome','len','HGT_count','extraneous_count', 'overlap_count']
    to_csv.append((csv_file,len(df), df[df['HGT']==1]['HGT'].count(),df[df['extraneous'] == True]['extraneous'].count(), df[(df['HGT']==1) & (df['extraneous'] == True)]['HGT'].count() ))
    
```
