# Is Firmicutes and protobacteria separable?

```python
protobacteria_merged['phylum'] = ['protobacteria' for i in range(len(protobacteria_merged))]
firmicutes_merged['phylum'] = ['firmicutes' for i in range(len(firmicutes_merged))]
```

```python
val_all_pd = pd.concat((protobacteria_merged,firmicutes_merged))
val_all_pd.reset_index(drop=True,inplace=True)
val_all_pd
```

```python
val_all = val_all_pd[['GC1 Mean', 'GC2 Mean', 'GC3 Mean', 'GCT Mean']].values
val_all
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(val_all) 
```

```python
x
```

```python
np.mean(x),np.std(x)
```

```python
from sklearn.decomposition import PCA
pca_HGTDB = PCA(n_components=2)
principalComponents_GCsignature = pca_HGTDB.fit_transform(x)
```

```python
principal_gc_Df = pd.DataFrame(data = principalComponents_GCsignature, columns = ['principal component 1', 'principal component 2'])
```

```python
principal_gc_Df
```

```python
import matplotlib.pyplot as plt
```

```python
indicesToKeep = val_all_pd['phylum'] == 'protobacteria'
indicesToKeep
```

```python
indicesToKeep = val_all_pd['phylum'] == 'firmicutes'
indicesToKeep
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['protobacteria', 'firmicutes']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = val_all_pd['phylum'] == target
    plt.scatter(principal_gc_Df.loc[indicesToKeep, 'principal component 1'], principal_gc_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)
    #plt.scatter(principal_breast_Df.loc[indicesToKeep.index, 'principal component 1'], principal_breast_Df.loc[indicesToKeep.index, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```
